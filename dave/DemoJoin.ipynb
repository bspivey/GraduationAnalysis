```python
import pandas as pd
pd.set_option("display.max_rows", 999)
pd.set_option("display.max_columns", 999)
pd.set_option('display.max_rows', 999)
pd.set_option('display.min_rows', 999)

%config IPCompleter.use_jedi = False
```

```python
grad = pd.read_csv('/Users/dave/src/CSE6242/LocalData/graduation_all.csv')
grad.columns = grad.columns.str.lower()
display(grad.shape)
display(grad.head())

grad['year'] = grad.srcyear.astype(str).str[:4]
display(grad.head())
grad = grad[grad.subgroup == 1.0]
grad.year = grad.year.astype('int')
display(grad.shape)
display(grad.year.value_counts())
display(grad.dtypes)
```

```python
apm = pd.read_csv("/Users/dave/src/CSE6242/LocalData/apm_all.csv")
display(apm.shape)
display(apm.dtypes)
display(apm.head())
```

```python
demo = pd.read_csv('/Users/dave/src/CSE6242/LocalData/Demographics_all.csv')
display(demo.shape)

display(demo.head())
```

```python
faru = pd.read_csv("/Users/dave/src/CSE6242/LocalData/FARU_all.csv")
# display(faru.dtypes)
display(faru.shape)
display(faru.head())
```

```python
df1 = pd.merge(apm, demo, 
               how='left', 
               left_on=["BEDSCODE", "YEAR"], 
               right_on=["DISTRICT", "YEAR"], 
               suffixes=["_apm", "_demo"])
display(df1.shape)
display(df1.head())
```

```python
df2 = pd.merge(df1, faru, 
              how='left', 
              left_on=["BEDSCODE", "YEAR"], 
              right_on=["DISTRICT", "YEAR"],
              suffixes=["_apm", "faru"])
df2.columns = df2.columns.str.lower()
display(df2.shape)
display(df2.head())
```

```python
# join grad info now
df = pd.merge(df2, grad,
                   how='left',
                   left_on = ['bedscode', 'year'], 
                   right_on = ['district_cd', 'year'], 
                   suffixes = ['_apm', '_grad'])
display(df.shape)
display(df.head())
display(df.dtypes)
```
