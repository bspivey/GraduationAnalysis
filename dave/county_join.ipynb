# County Name Map
* First we load the county to district name mapping from 36_County_Coverage_A.txt (from Census.gov)
* Second, we load the district name to county mapping from grad_rate.csv District Aggregation (NYSED)
* Third, we load the district_cd to district name mapping from BasicInfo.csv (from Cornell)
* Then we use that aggregate table to join the county on the school district name

## County Coverage (cc)
mapping school district name to county name

```python
# display(district.head())
county_coverage = pd.read_table("~/src/CSE6242/dave/36_County_Coverage_A.txt", sep="[ ]{2,}", header=0)
county_coverage['district_name'] = county_coverage.name.str.lower()
cc = county_coverage[["county_name", "district_name"]]
display(cc.head())
display(cc.dtypes)
display(cc.district_name.nunique())
```

## Grad County Name (gc)
Do we even need this?

```python
df_grad = pd.read_pickle("../DataAll/df_grad_rate_all.pkl")
df_grad.columns = df_grad.columns.str.lower()
counties = df_grad[df_grad.aggregation_type=='County']
districts = df_grad[df_grad.aggregation_type=='District']
districts.head()
district_to_county_map = districts[["aggregation_name", "county_name", "county_code"]]
district_to_county_map['aggregation_name'] = district_to_county_map.aggregation_name.str.lower()
district_to_county_map['county_name'] = district_to_county_map.county_name.str.lower()
district_to_county_map["district_shortname"] = district_to_county_map["aggregation_name"].str.rsplit(" ", n=2).str[:-2].str[0]
display(district_to_county_map.shape)
gc = district_to_county_map.drop_duplicates()[['district_shortname', 'county_name']]
display(gc.shape)
display(gc.head())
```

## District Code to District Name Mapping
from BasicInfo.csv (Cornell)

```python
basicinfo = pd.read_csv("../LocalData/BasicInfo2.csv")
basicinfo.columns = basicinfo.columns.str.lower()
basicinfo["district_name"] = basicinfo.district_name.str.lower()
basicinfo['name'] = basicinfo["district_name"].str.replace(" sd", "")
basicinfo['name'] = basicinfo["name"].str.replace(" csd", " central")
basicinfo['name'] = basicinfo["name"].str.replace(" ufsd", "")

b = basicinfo[["districtid", "name"]]
display(b.head())
b.districtid.count()
```

## Merge the County Name to District Code (Final Mapping Step)

```python
# how many district IDs are there?
display(b.districtid.nunique())

# map1 = using cc
map1 = pd.merge(b, cc, how="inner", left_on="name", right_on="district_name")
display(map1.head())
display(map1.shape)

# map2 = using gc
map2 = pd.merge(b, gc, how="inner", left_on="name", right_on="district_shortname")
display(map2.head())
display(map2.shape)

```

# Left join County Name to District DF

```python
df.head()
```

```python
df_with_counties = pd.merge(df, map1,
                           how='left',
                           left_on='district_cd',
                           right_on='districtid',
                           suffixes=['_from_df', '_from_map'])
```
